#### Elements of a plot

In Matplotlib, the figure (an instance of the class plt.Figure) can be thought of as a single container that contains all the objects representing axes, graphics, text, and labels.  

The axes (an instance of the class plt.Axes) is what we see above: a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization

```python
plt.figure(figsize = (8, 6))

plt.plot(age, dev_y, color = 'gray', ls = '--', marker = 'o', label = 'Overall Salary')
# if you write 2 or more "plt.plot"s in the same cell, it creates 
# multiple plots on the same axes

plt.plot(age, py_dev_y, color = '#33FF52', lw = 3, label = 'Python Salary')

# add label
plt.legend(loc = 'upper left')

# add title
plt.title('Overall Salary in USD')

# add x and y labels
plt.xlabel('Age')
plt.ylabel('Salary in USD')

# add gridlines
plt.grid(ls = '--')

# add annotations
plt.annotate('This salary', xy = (25,38000), xytext = (28, 42000), arrowprops = dict(facecolor = 'gray'))
```

### Subplots: multiple axes

- Multiple axes (plots) on a single figure
- Use plt.subplot(num_rows, num_cols, index, **kwargs)



```python
# Within a figure - create two subplots( one row and two columns)
plt.figure(figsize = (8,6))

plt.subplot(121)
#(row,column,which plot)
plt.text(0.5, 0.5, str((121)))


plt.subplot(122)
plt.text(0.5, 0.5, str((122)))
```

```python
# plot different mathematical functions vs x in a single figure
x = [x * 0.25 for x in range(-25,26)]
pi = math.pi

y1 = [math.sin(x) for x in x]
y2 = [math.cos(x) for x in x]
y3 = [math.sin(x) * -1 for x in x]
y4 = [math.cos(x) * -1 for x in x]

plt.figure(figsize = (8,8))
plt.suptitle('Some Math Functions')

plt.subplot(221)
plt.plot(x, y1, color = 'r', ls = '--')
plt.title('Sine Function')


plt.subplot(222)
plt.plot(x, y2, color = 'g', ls = '--', marker = '^')

plt.subplot(223)
plt.plot(x, y3, color = 'b', ls = '--', marker = 'o')

plt.subplot(224)
plt.plot(x, y4, color = 'm', ls = '--', marker = 'v')
plt.axhline(0, color = 'm')
plt.fill_between(x, y4, color = 'm', alpha = 0.9)
# alpha = controls transparency

plt.tight_layout()

plt.show()
# only needed for VS Code, otherwise it will not print plots
# Jupyter runs this command automatically
```

### Histograms: explore distribution of a sample/population

Explore the distribution of a sample/population

```python
# Explore age of respondents from stackoverflow survey

data = pd.read_csv('https://raw.githubusercontent.com/ravinder79/data/main/age_data.csv', index_col=0)
```

```python
# %matplotlib inline
data
```

```python
# %matplotlib inline
plt.hist(data, bins = range(0,100,5));
# third value is step size
# in this case it changes the width of each bar
```

```python
plt.figure(figsize = (8, 6))

x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2,1, 1000)
x3 = np.random.normal(3, 1.5, 1000)
x4 = np.random.normal(7, 0.6, 1000)


# make subplots using for loop
plt.figure(figsize = (8,6))
x = [x1, x2, x3, x4] # make a list of lists to iterate through

for i in range(0, len(x)):
    plt.subplot(2,2,i+1)
    plt.hist(x[i], color = 'w', ec = 'k')
```

### Summary

Pyplot - Create a plot
- Line plot: plt.plot()
- Scatter plot: plt.scatter()
- Histogram: plt.hist()
- Bar plot: plt.bar()


Pyplot - Customize the plot
- Add title: plt.title()
- Add x or y label: plt.xlabel() or plt.ylabel()
- Add gridlines: plt.grid()
- Add legend: plt.legend()
- Set x and y axis limits: plt.xlim(0,10) or plt.ylim()
- Customize x or y ticks: plt.xticks(), plt.yticks()
- Annotate the plot: plt.annotate()
- Set figure size: plt.figure(figsize=(width, height))
- Draw a horizontal line across axes: plt.axhline()
- Draw a verticle line across axes: plt.axvline()
- Add text on a plot: plt.text()
- plot spacing: plt.tight_layout()

Pyplot - Create Subplots
- plt.subplot(num_rows, num_cols, index, **kwargs)
